```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
plt.style.use('ggplot')
attendees = pd.read_csv(filepath, index_col=False).drop_duplicates('Email')
```

```python
attendance = np.array([len(attendees[attendees['Attended'] == 'Yes']),
            len(attendees[attendees['Attended'] == 'No'])]) #[attendees, nonattendees]
percentage = np.array(list(map(lambda x: x / sum(attendance), attendance)))
attend_df = pd.DataFrame(np.vstack((attendance, percentage)).T, 
                         columns=['Attendance', '%'],index=['Attendees', 'Non-Attendees'])
attend_df
```

```python
attend_df['Attendance'].plot.pie(y='Attendance', autopct="(%1.1f)%%", pctdistance=0.6, labeldistance=1.1)
plt.show()
```
